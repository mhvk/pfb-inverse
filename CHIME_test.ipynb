```python
%pylab inline
from pfb import pfb, inverse_pfb
import numpy as np
```

```python
ntime = 2**25

# Generate a white noise timestream
ts = np.random.standard_normal(ntime)

# Perform the PFB
spec_pfb = pfb(ts, 1025)

# Perform the inverse
rts = inverse_pfb(spec_pfb, 4)
```

```python
# show a small section
imshow(spec_pfb[:300].real, interpolation='nearest', aspect='auto', cmap='RdBu')
colorbar()
```

```python
block_res = ts.reshape(-1, 2048) - rts

title("Standard deviation of residuals")
semilogy(block_res.std(axis=1))
xlabel("Blocks")
```
